# Pandas Plot

### 1. Depict Market and Type wise sum of Budget Profit on a bar chart

```python
df.head(2)
```

```python
b=df.groupby(['Market','Type'])['Budget Profit'].sum()
```

```python
b.plot(kind='bar',color='cyan',figsize=(15,5))
plt.title('Market and Type wise sum of Budget Profit')
plt.xlabel('Market & Type')
plt.ylabel('Budget Profit')
plt.xticks(rotation=80)
plt.show()
```

 ### 2 Depict Product Type and State wise mean of Sales on a area chart

```python
df['Product Type'].value_counts()
```

```python
df['State'].value_counts()
```

```python
pts=df.groupby(['Product Type','State'])['Sales'].mean()
pts
```

```python
pts.plot(kind='area',color='black',figsize=(10,5))
plt.xlabel('Product type & State')
plt.ylabel('Sales')
plt.xticks(rotation=30)
plt.grid()
plt.show()
```

 ### 3. Depict Product and Type wise sum of Budget COGS on a line chart

```python
b=df.groupby(['Product','Type'])['Budget COGS'].sum()
b
```

```python
plt.figure(figsize=(10,5))
b.plot(kind='line',color='cyan',marker='*')
plt.xlabel('Product & type')
plt.ylabel('Budget COGS')
plt.xticks(rotation=80)
plt.grid()
plt.show()
```

### 4 Depict Product and Type wise sum of Budget Profit and Budget Sales on a line Chart

```python
pt=df.groupby(['Product','Type'])['Budget Profit','Budget Sales'].sum()
pt
```

```python
pt.plot(kind='line',marker='*', figsize=(10,5))
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

### 5 Depict Market and Product Type based mean of Sales in a bar char and on a stacked bar chart seperately

```python
bsc=df.groupby(['Market','Product Type'])['Sales'].mean()
bsc
```

```python
bsc.plot(kind='bar',color='cyan', figsize=(10,5))
plt.xlabel('Market & Product Type')
plt.ylabel('Sales')
plt.title('Market & Product Type based on mean of Sales')
plt.xticks(rotation = 30)
plt.show()
```

```python
unbse=bsc.unstack()
unbse
```

```python
unbse.plot(kind='bar', figsize=(10,5))
plt.xlabel('Market & product Type')
plt.ylabel('sales')
plt.grid()
plt.title('Market Product Type vs Sales Means')
plt.xticks(rotation='30')
plt.legend()
plt.show()
```

```python
unbse.plot(kind='bar',stacked = True, figsize=(15,5))
plt.grid()
plt.xticks(rotation='30')
plt.title('Market Product Type vs Sales Means on Stacked bar chart')
plt.show()
```

### 6. Find State and Market size based mean of Sales and COGS where Market is not Central and Profit is over 350, Depict the result on a Line Chart, Stacked bar and area chart separately

```python
mcp=df[(df['Market']!='Central') & (df['Profit']>360)]
mcp.shape
mcp
```

```python
msc=mcp.groupby(['Market Size','State'])['Sales','COGS'].mean()
msc
```

```python
msc.plot(kind='line',marker='*', figsize=(10,5))
plt.xticks(rotation=90)
plt.title('State, Market size & mean of Sales')
plt.show()
```
