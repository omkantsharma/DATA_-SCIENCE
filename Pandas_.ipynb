# Merge

<b>inner : </b> Return only the rows in which the left table has matching keys in the right table


```python
# Inner  
d1 = pd.merge(df1, df2, on='Customer_id', how='inner')
# d1[['Product','State']]
d1
```

## <b>outer :</b> Returns all rows from both tables, join records from the left which have matching keys in the right table.

```python
pd.merge(df1, df2, on='Customer_id', how='outer')
```

<b>Left outer : </b> Return all rows from the left table, and any rows with matching keys from the right table.

```python
pd.merge(df1, df2, on='Customer_id', how='left')
```

<b>Right outer :</b> Return all rows from the right table, and any rows with matching keys from the left table.

```python
import pandas as pd
import numpy as np
```

```python
pd.merge(df1, df2, on='Customer_id', how='right')
```

<br>

<ol>

# Joins

```python
df_a={
    'subject_id':[1,2,3,4,5],
    'first_name':['Ajay','Abhi','Aman','Avi','Aksh'],
    'last_name':['Anderson','Ackerman','all','Aoni','Aliches']
}
df_b={
    'subject_id':[4,5,6,7,8],
    'first_name':['Billy','Navi','Swati','Shivali','Kamal'],
    'last_name':['Bonder','Black','Balwner','Nrie','Ntise']
}
df_a=pd.DataFrame(df_a)
df_b=pd.DataFrame(df_b)
df_a,df_b
```

```python
df_a
```

```python
df_b
```

```python
df_concat=pd.concat([df_a,df_b])
df_concat
```

```python
df_concat_axis1=pd.concat([df_b,df_a],axis=1)
df_concat_axis1
```

```python
# merge with right join
pd.merge(df_a,df_b,on='subject_id',how='right')
```

```python
pd.merge(df_b,df_a,on='subject_id',how='right')
```

```python
'''
Merge with left join 
left outer join product a complete set of records form
Table A, with the matching records (where available) in 
Table B, if there is no match, the right side will contain
nulll."

'''
pd.merge(df_a,df_b,on='subject_id',how='left')
```

```python
pd.merge(df_a,df_b,on='subject_id',how='inner')
```

```python
pd.merge(df_a,df_b,on='subject_id',how='outer')
```
